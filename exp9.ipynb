## Perform CRUD operations on a list storing student marks.

```python
students = []

while True:
    print("\n--- Student Marks Management ---")
    print("1. Add Student Marks (Create)")
    print("2. View All Marks (Read)")
    print("3. Update Student Marks (Update)")
    print("4. Delete Student Record (Delete)")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    # CREATE
    if choice == '1':
        name = input("Enter student name: ")
        marks = float(input("Enter marks: "))
        students.append({"name": name, "marks": marks})
        print("Student added successfully!")

    # READ
    elif choice == '2':
        if not students:
            print("No records found.")
        else:
            print("\nStudent Records:")
            for i, student in enumerate(students):
                print(f"{i}. {student['name']} - {student['marks']}")

    # UPDATE
    elif choice == '3':
        index = int(input("Enter student index to update: "))
        if 0 <= index < len(students):
            new_marks = float(input("Enter new marks: "))
            students[index]["marks"] = new_marks
            print("Marks updated successfully!")
        else:
            print("Invalid index!")

    # DELETE
    elif choice == '4':
        index = int(input("Enter student index to delete: "))
        if 0 <= index < len(students):
            removed = students.pop(index)
            print(f"{removed['name']} deleted successfully!")
        else:
            print("Invalid index!")

    # EXIT
    elif choice == '5':
        print("Exiting program.")
        break

    else:
        print("Invalid choice! Please try again.")
```
