## Sort employee records based on salary.

```python
employees = {
    "101": {"Name": "Ravi", "Salary": 25000},
    "102": {"Name": "Neha", "Salary": 30000},
    "103": {"Name": "Amit", "Salary": 28000}
}

# Sorting employees by salary (ascending order)
sorted_employees = sorted(employees.items(), key=lambda x: x[1]["Salary"])

print("Employees sorted by salary:\n")

for emp_id, details in sorted_employees:
    print(f"ID: {emp_id}, Name: {details['Name']}, Salary: {details['Salary']}")
```
