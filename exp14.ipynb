## Write a program to count vowels, consonants, digits, and special characters in a string.

```python
# Input string
text = input("Enter a string: ")

vowels = "aeiouAEIOU"
v_count = c_count = d_count = s_count = 0

for ch in text:
    if ch.isalpha():  # Check if alphabet
        if ch in vowels:
            v_count += 1
        else:
            c_count += 1
    elif ch.isdigit():  # Check if digit
        d_count += 1
    else:  # Special character
        s_count += 1

# Output results
print("Vowels:", v_count)
print("Consonants:", c_count)
print("Digits:", d_count)
print("Special characters:", s_count)
```
