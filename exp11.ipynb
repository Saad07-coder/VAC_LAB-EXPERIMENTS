## Write a program to search employee details using employee ID.

```python
employees = {
    "101": {"Name": "Ravi", "Salary": 25000},
    "102": {"Name": "Neha", "Salary": 30000},
    "103": {"Name": "Amit", "Salary": 28000}
}

# Search employee by ID
emp_id = input("Enter Employee ID to search: ")

if emp_id in employees:
    print("Employee Found!")
    print("ID:", emp_id)
    print("Name:", employees[emp_id]["Name"])
    print("Salary:", employees[emp_id]["Salary"])
else:
    print("Employee not found.")
```
